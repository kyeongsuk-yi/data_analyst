## 1. 사전준비

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

pd.set_option('expand_frame_repr', False)
```

```python
X_test = pd.read_csv("./yemoonsaBigdata/datasets/Part2/mpg_X_test.csv")
X_train = pd.read_csv("./yemoonsaBigdata/datasets/Part2/mpg_X_train.csv")
y_train = pd.read_csv("./yemoonsaBigdata/datasets/Part2/mpg_y_train.csv")
```

```python
# fillna 함수 
# sklearn - SimpleImputer

from sklearn.impute import SimpleImputer 

# imp = SimpleImputer()
imp = SimpleImputer(missing_values=np.nan, strategy='mean') # median, most_frequent, constant (fill_value=0)

X_train[['horsepower']] =  imp.fit_transform(X_train[['horsepower']])
X_test[['horsepower']] =  imp.fit_transform(X_test[['horsepower']])
```

```python
del_cols = ['name']
num_cols = [ 'mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
cat_cols = []
y_col = []
```

```python
X_train.drop(del_cols, axis=1, inplace=True)
X_test.drop(del_cols, axis=1, inplace=True)
```

## 3. 데이터 모형 구축 

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, stratify=y_train, test_size=0.2, random_state=55) #stratify=y_train[y_col], 
```

```python
print(y_tr.value_counts()/len(y_tr))
print(y_val.value_counts()/len(y_val))
```

```python
y_train.value_counts() /len(y_train)
```

```python
## 데이터 스케일링

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X_tr[num_cols])
X_tr[num_cols] = sc.transform(X_tr[num_cols])
X_val[num_cols] = sc.transform(X_val[num_cols])
X_test[num_cols] = sc.transform(X_test[num_cols])
```

```python
X_tr
```

## 3. 데이터 모형 구축

```python

# KNN, Decision Tree
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_tr, y_tr)

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=10, random_state=55)
dt.fit(X_tr, y_tr)
```

## 4. 데이터 모형 평가

```python
from sklearn.metrics import roc_auc_score

y_tr_pred_knn = knn.predict_proba(X_tr)[:, 1]
y_tr_pred_dt = dt.predict_proba(X_tr)[:, 1]

y_val_pred_knn = knn.predict_proba(X_val)[:, 1]
y_val_pred_dt = dt.predict_proba(X_val)[:, 1]
```

```python
print(f'knn: {roc_auc_score(y_tr, y_tr_pred_knn).round(3)}, / {roc_auc_score(y_val, y_val_pred_knn).round(3)} ')
print(f'dt: {roc_auc_score(y_tr, y_tr_pred_dt).round(3)}, / {roc_auc_score(y_val, y_val_pred_dt).round(3)} ')
```

```python
for i in range(2, 16):
  
  model = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
  model.fit(X_tr, y_tr)

  y_tr_pred = model.predict_proba(X_tr)[:,1]
  y_val_pred = model.predict_proba(X_val)[:,1]

  score_tr = roc_auc_score(y_tr, y_tr_pred)
  score_val = roc_auc_score(y_val, y_val_pred)
  print(f'{i}개의 이웃 확인 : {round(score_tr, 3)} / {round(score_val, 3)}')
```
